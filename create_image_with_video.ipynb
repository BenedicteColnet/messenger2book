Create frame with video or GIFS

```python
import os
import cv2 # need a broader package
from PIL import Image # need Image module
import re
```

```python
messenger_gifs = os.listdir("data/messenger/gifs")
messenger_gifs
```

```python
for gif in messenger_gifs:
    im = Image.open("data/messenger/gifs/"+gif)
    im.seek(0)
    res = re.search("([0-9]*)_([0-9]*)_([0-9]*)_(.).", gif)
    new_name = f"{res.group(1)}_{res.group(2)}_{res.group(3)}_{res.group(4)}"
    im.save("data/messenger/gifs_as_frame/"+new_name+".png")
```

```python
messenger_video = os.listdir("data/messenger/videos")
messenger_video
```

```python
def getFirstFrame(videofile):
    vidcap = cv2.VideoCapture(videofile)
    success, image = vidcap.read()
    if success:
        new_name = videofile[:-3]+"jpg"
        print(new_name)
        cv2.imwrite("data/telegram/ChatExport_2020-11-15/video_files_as_frame/new_name", image)  # save frame as JPEG file
        
        
for video in messenger_video:
    if ".mp4" in video:
        vidcap = cv2.VideoCapture("data/messenger/videos/" + video)
        success, image = vidcap.read()
        if success:
            new_name = video[5:-4]+'.png'
            print(new_name)
            cv2.imwrite("data/messenger/videos_as_frame/"+new_name, image)
   
```
